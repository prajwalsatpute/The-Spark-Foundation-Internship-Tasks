**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**Getting Data**

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

**As we have** ***species*** **column as our target column. But** **before assigning target values we need to convert it into numerical form.**

```python
iris['species'] = iris['species'].replace({'setosa' : 0,'versicolor' : 1,'virginica' : 2})
iris['species'].values
```

**Now we will create our feature variable as X and target variable as y and split our data as training and testing data.**

```python
X = iris.drop('species',axis=1)
y = iris['species']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

**Let's create our Decision Tree Algorithm**

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
print('Decision Tree Classifier Created')
```

```python
pred = dtree.predict(X_test)
```

**Compare our predicted data with actual data**

```python
dfc = pd.DataFrame({'Actal Data' : y_test , 'Predicted Data' : pred})
dfc.head()
```

**Now the second task is that to visualize this classifier graphically. For that we need some libraries. Let's import them.**

```python
from sklearn.tree import export_graphviz
from sklearn import tree
from sklearn.tree import plot_tree
```

**Visualizing the classifier.**

```python
plt.figure(figsize=(15,15))
plot_tree(dtree,feature_names=iris.columns,class_names=["Iris-setosa","Iris-versicolor","Iris-virginica"],
          filled=True,precision=4,rounded=True)
```

**We have done required task.**

**At last let's check our classifier's accuracy.**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

**Cool , We have got 100% accuracy of our classifier.**
